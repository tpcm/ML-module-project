```python
# Read in all datasets csv as panda Data Frames
training = pd.read_csv('/kaggle/input/i-wish-i-was-there/training.csv', index_col='ID')
testing =pd.read_csv('/kaggle/input/i-wish-i-was-there/testing.csv', index_col='ID')
confidence = pd.read_csv('/kaggle/input/i-wish-i-was-there/annotation_confidence.csv', index_col='ID')
additional_training = pd.read_csv('/kaggle/input/i-wish-i-was-there/additional_training.csv', index_col='ID')
```

```python
# Import all necessary modules
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.impute import SimpleImputer
```

```python
# Merge training and additional training data
merged_training = pd.concat([training, additional_training])
# Separate predictions
merge_no_predict = merged_training.drop(['prediction'], axis=1)
# Perform Imputation on training data, use MOST FREQUENT value
imputer = SimpleImputer(missing_values = np.nan, strategy='most_frequent')
# Imputer is fitted n testing data
imputer.fit(testing)
X = imputer.transform(merged_training.drop(['prediction'], axis=1))
X = pd.DataFrame(X, columns=testing.columns)
# Assign predictions to y
y = pd.DataFrame(merged_training['prediction'])
display(X)
display(y)
```

```python
# Create testing data class proportions dictionary to pas into model
class_weights = {0:0.4431, 1:0.5569}
# Instantiate model with probability set to True
model = SVC(kernel='rbf', gamma='auto', C=3.5, class_weight=class_weights, probability=True)
# Fit model on the training data and predictions
model.fit(X,y.values.ravel())
# Calculate classification probability estimates
clp = model.predict_proba(testing)
# Make predictions
lab = model.predict(testing)
```

```python
# Get max probability from class estimates, this is used as the prediction confidence
pred_conf = [max(p) for p in clp]
# Augment confidence annotations provided with class predictions confidence
total_conf = np.append(confidence['confidence'],pred_conf)
```

```python
# Augment training labels with testing data predictions
y_plus = pd.concat([y, pd.DataFrame(lab, columns= y.columns)], ignore_index=True)
y_plus
```

```python
# Augment training data with testing data
# This completes the transformation of the problme into a semi-supervised one
new_train = pd.concat([X, testing], ignore_index=True)
new_train.isnull().sum()
```

```python
# Perform feature selection on the augmented training data
selector_2 = SelectKBest(score_func=chi2, k=1500)
selector_2 = selector_2.fit(new_train, y_plus)
final_K_train = selector_2.transform(new_train)
```

```python
# Instantiate final model to be used in final predictions on testing data
# No class weights nor, and probability left out to speed up
final_model = model = SVC(kernel='rbf', gamma='auto', C=4.5)
final_model.fit(final_K_train, y_plus.values.ravel(), total_conf)
```

```python
# Transform testing data into Kbest features
final_test = selector_2.transform(testing)
```

```python
# Make final predictions on testing data and convert to csv file
predictions = model.predict(final_test)
df_predictions = pd.DataFrame(predictions, index=testing.index, columns=['prediction'])
df_predictions.to_csv('submit.csv')
df_predictions
```
